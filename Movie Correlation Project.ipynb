```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configration of the plots we will create


# Read in the data

df = pd.read_csv(r'C:\Users\D.C - L\Desktop\PORTFOLIO BUILD\movies.csv')

```

```python
# Let's look at the data

df.head()
```

```python
# Looking if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# We have some missing values. Just for this project, we are going to drop them

df = df.dropna()
```

```python
# We are now going to check the data types for our columns

df.dtypes
```

```python
# We are going to change the datatype for column budget and gross to int64

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# Let's check our dataset if the changes applied on those columns

df.head()
```

```python
# Extracting correct year from "released" column and adding new column "yearcorrect" for those values

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Let's check if the new column is added with the extracted year

df.head()
```

```python
# Let's start sorting our data by "gross" column in descending order

df_sorted = df.sort_values(by='gross', ascending=False)
top_records = df_sorted.head(10)  # Select the top 10 records

top_records.head()
```

```python
# If we want to view all of the data, we can run this

pd.set_option('display.max_rows', None)
```

```python
# Let's start looking at correlation
# different types of correlation pearson, kendall, spearman

df.corr(method='pearson', numeric_only=True)
```

```python
# High correlation between budget and gross
```

```python
# We can visualize this to see the correlation matrix

correlation_matrix = df.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()

# Anything that shows with bright colors got high correlation while darker means low correlation
```

```python
# Some of our columns is not numeric, but we can create a unique numeric representation for them

df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtypes == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
# Now we can use our matrix to find correlation in any single field in our data, 
# because all of the values have numeric representation

correlation_matrix = df_numerized.corr(method='pearson', numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()

# This shows that the budget and votes have high correlation to gross income
```

```python
df_numerized.corr()
```

```python
# We can also use this to quickly see the correlation

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# We can also see the correlation using this method

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# Now that we have the sorted pairs, we can now look at the sorted pairs that's greater than 0.5

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
